```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Wed Jun  5 09:17:53 2019

@author: chairat
"""
import numpy as np
import cv2
import imutils
import cv2
import time
from gaze_tracking import GazeTracking
from random import randint
gaze = GazeTracking()
webcam = cv2.VideoCapture(0)

#import matplotlib.image as mpimg


img = cv2.imread('./menu/L1M1.png',cv2.IMREAD_COLOR)

#cv2.namedWindow("output", cv2.WINDOW_NORMAL)
#imS = cv2.resize(img, (1024, 640))
#imSS = imS.copy();
#cv2.imshow('image',img)



#============ initial system config =================
nextNemu = 'L1M1'
i=0
countDict = {'cLeft':0, 'cRight': 0, 'cCenter': 0, 'cBlinking': 0}
#selecedtDict = {'cLeft':0, 'cRight': 0, 'cCenter': 0, 'cBlinking': 0}
# flowDict = {'lv1Left': 'menuLV2L' , 'lv1Right': 'menuLV2R' , 'lv1Center': 'menuLV2C'}
curPosition = 'L1M1'
selectThreshold = 20
menuDict = {'L1M1' : './menu/L1M1.png', 'L2M1' : './menu/L2M1.png', 'L2M2' : './menu/L2M2.png', 'L2M3' : './menu/L2M3.png'
           
           , 'L2MF' : './menu/L2MF.png'};

focusmenuChange = 'false'
menuChange = 'false'

#====================================================

def runModel():
    
    return model2()



def model1():
    
    
    global cv2
    
    
        # We get a new frame from the webcam
    _, frame = webcam.read()

    # We send this frame to GazeTracking to analyze it
    frame = Zoom(frame,1)
    
    gaze.refresh(frame)

    frame = gaze.annotated_frame()
    text = ""
    eyePo = 0

    if gaze.is_blinking():
        text = "Blinking"
        eyePo = 0
    elif gaze.is_left():
        text = "Looking left"
        eyePo = 1
    elif gaze.is_center():
        text = "Looking center"
        eyePo = 2
    elif gaze.is_right():
        text = "Looking right"
        eyePo = 3
    

    cv2.putText(frame, text, (90, 60), cv2.FONT_HERSHEY_DUPLEX, 1.6, (147, 58, 31), 2)

    left_pupil = gaze.pupil_left_coords()
    right_pupil = gaze.pupil_right_coords()
    cv2.putText(frame, "Left pupil:  " + str(left_pupil), (90, 130), cv2.FONT_HERSHEY_DUPLEX, 0.9, (147, 58, 31), 1)
    cv2.putText(frame, "Right pupil: " + str(right_pupil), (90, 165), cv2.FONT_HERSHEY_DUPLEX, 0.9, (147, 58, 31), 1)
    cv2.imshow("Demo", frame)
    print (text)
    
    return eyePo

def model2():
    eyePo = 0
    eyePo = randint(1, 3)
    print ('model 1 return : ',str(eyePo))
    return eyePo


def focusLeft():
    cv2.rectangle(img,(40,161),(385,456),(0,255,0),15)
    return 1

def focusRight():
    cv2.rectangle(img,(1215,161),(1560,456),(0,255,0),15)
    return 1

def focusCenter():
    cv2.rectangle(img,(626,161),(972,456),(0,255,0),15)
    return 1

def selected(countKey):
    #TODO 
    global countDict
    global menuChange
    global nextNemu
    menuChange = 'false'
    if(countDict[countKey] >= selectThreshold):
        menuChange = 'true'
        if curPosition == 'L1M1':
            if countKey == 'cLeft':
                nextNemu = 'L2M1'            
            elif countKey == 'cCenter':
                nextNemu = 'L2M2'
            elif countKey == 'cRight':
                nextNemu = 'L2M3'
                
            print ('nextNemu : ' + menuDict[nextNemu])
            countDict = countDict.fromkeys(countDict, 0)
            
        elif curPosition == 'L2M1':
            if countKey == 'cLeft':
                nextNemu = 'L1M1'            
            elif countKey == 'cCenter':
                nextNemu = 'L2MF'
            elif countKey == 'cRight':
                nextNemu = 'L1M1'
            
            print ('nextNemu : ' + menuDict[nextNemu])
            countDict = countDict.fromkeys(countDict, 0)
            
        elif curPosition == 'L2M2':
            if countKey == 'cLeft':
                nextNemu = 'L1M1'            
            elif countKey == 'cCenter':
                nextNemu = 'L2MF'
            elif countKey == 'cRight':
                nextNemu = 'L1M1'
                
        elif curPosition == 'L2M3':
            if countKey == 'cLeft':
                nextNemu = 'L1M1'            
            elif countKey == 'cCenter':
                nextNemu = 'L2MF'
            elif countKey == 'cRight':
                nextNemu = 'L1M1'
        elif curPosition == 'L2MF':
            if countKey == 'cLeft':
                nextNemu = 'L1M1'            
            elif countKey == 'cCenter':
                nextNemu = 'L1M1'
            elif countKey == 'cRight':
                nextNemu = 'L1M1'
            
            print ('nextNemu : ' + menuDict[nextNemu])
            countDict = countDict.fromkeys(countDict, 0)

    
    return 1

def Zoom(cv2Object, zoomSize):
    # Resizes the image/video frame to the specified amount of "zoomSize".
    # A zoomSize of "2", for example, will double the canvas size
    cv2Object = imutils.resize(cv2Object, width=(zoomSize * cv2Object.shape[1]))
    # center is simply half of the height & width (y/2,x/2)
    center = (cv2Object.shape[0]/2,cv2Object.shape[1]/2)
    # cropScale represents the top left corner of the cropped frame (y/x)
    cropScale = (center[0]/zoomSize, center[1]/zoomSize)
    # The image/video frame is cropped to the center with a size of the original picture
    # image[y1:y2,x1:x2] is used to iterate and grab a portion of an image
    # (y1,x1) is the top left corner and (y2,x1) is the bottom right corner of new cropped frame.
    cv2Object = cv2Object[cropScale[0]:(center[0] + cropScale[0]), cropScale[1]:(center[1] + cropScale[1])]
    return cv2Object

def main():
    global img
    global nextNemu
    global menuDict
    global curPosition
    while True:
        print(str(menuDict[nextNemu]))
        curPosition = nextNemu
        img = cv2.imread(menuDict[nextNemu],cv2.IMREAD_COLOR)
        print('Reset IMG')
        #    #img = cv2.imread(menuDict[nextNemu],cv2.IMREAD_COLOR)
        
        eyePosition = runModel()


        i = eyePosition;

        if i==1:
            #R1
            countDict['cLeft'] +=1
            focusLeft()
            selected('cLeft')
            print('cLeft')

        elif  i==2:
            #R2
            countDict['cCenter'] +=1
            focusCenter()
            selected('cCenter')
            print('cCenter')

        elif  i==3:
            #R3
            countDict['cRight'] +=1
            focusRight()
            selected('cRight')
            print('cRight')


        elif  i==0:
            #Blink
            countDict['cBlinking'] +=1
            print('Blink')

        else:
            print(i)


        
        cv2.imshow('Menu : ',img)
        print(countDict)
        time.sleep(0)



        if cv2.waitKey(1) == 27:
            break

if __name__ == '__main__':
    
    main()
    #cv2.imshow('Menu : ',img)
```
